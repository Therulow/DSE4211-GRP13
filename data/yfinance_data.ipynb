```python
import yfinance as yf
import pandas_market_calendars as mcal
import pandas as pd

# Original tickers
tickers = {
    "GC=F": "gold",
    "SI=F": "silver",
    "BZ=F": "brent",
    "^DJI": "dji",
    "^SPX": "spx",
    "^RUT": "rut",
    "^IXIC": "nasdaq",
    "CNY=X": "usdcny",
    "EUR=X": "usdeur",
    "^VIX": "vix"
}

start_date = "2016-02-01"
end_date = "2026-02-01"

# Get NYSE calendar
nyse = mcal.get_calendar('NYSE')

# Get valid trading days
nyse_schedule = nyse.schedule(start_date=start_date, end_date=end_date)

nyse_trading_days = nyse_schedule.index

# Download data
data = yf.download(
    list(tickers.keys()),
    start=start_date,
    end=end_date,
    interval="1d",
    auto_adjust=False
)

# Build single dataframe
df = pd.DataFrame(index=data.index)

df = df.loc[df.index.isin(nyse_trading_days)]

for raw_ticker, clean_name in tickers.items():
    df[f"{clean_name}_close"] = data["Adj Close"][raw_ticker]
    df[f"{clean_name}_vol"] = data["Volume"][raw_ticker]

print(df.head(10))

# Removing volume columns for FX rates as data is unavailabe 
df = df.drop(columns=["usdcny_vol", "usdeur_vol", "vix_vol"])
```

```python
df.to_csv("macro_indicators_data_2016_2026.csv")
```
